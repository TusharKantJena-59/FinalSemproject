```python
# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 3
seg_img_X=[]

for pic in X:
  Z= pic.reshape((-1,3)) #2d
  # conversion to np.float32
  Z = np.float32(Z)
  ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS) #attempts=10
# Convert back shape of original image
  center = np.uint8(center)
  res = center[label.flatten()]
  out2 = res.reshape((pic.shape))
  seg_img_X.append(out2)
i=0

os.mkdir('/content/drive/MyDrive/Brain Tumor Localization Project/segmented/no')
for seg_pic in seg_img_X:
  cv2.imwrite("/content/drive/MyDrive/Brain Tumor Localization Project/segmented/no/segK3_NO%i.jpg" %i, seg_pic)
  i+=1
  plt.figure()
  plt.imshow(seg_pic)
```

```python
contrast = []
energy =  []
homogeneity =  []
correlation =  []
data=[]

current_path ='/content/drive/MyDrive/Brain Tumor Localization Project/segmented/no'

for file in os.listdir(current_path):
  path = (current_path + "/" + file)
  img = skimage.io.imread(path, as_gray=True)
  img = skimage.img_as_ubyte(img)
  img = np.asarray(img, dtype="int32")
            
  g = skimage.feature.greycomatrix(img, [1], [0], levels=img.max()+1, symmetric=False, normed=True)
  glcm_contrast = skimage.feature.greycoprops(g, 'contrast')[0][0]
  glcm_energy = skimage.feature.greycoprops(g, 'energy')[0][0]
  glcm_homogeneity = skimage.feature.greycoprops(g, 'homogeneity')[0][0]
  glcm_correlation = skimage.feature.greycoprops(g, 'correlation')[0][0]
  contrast.append(glcm_contrast)
  energy.append(glcm_energy)
  homogeneity.append(glcm_homogeneity)
  correlation.append(glcm_correlation)
```

```python
Categories=['no','yes']
flat_data_arr=[] #input array
target_arr=[] #output array
datadir='/content/drive/MyDrive/Brain Tumor Localization Project/Brain_tumor_detection_dataset'
#path which contains all the categories of images
for i in Categories:
	
	print(f'loading... category : {i}')
	path=os.path.join(datadir,i)
	for img in os.listdir(path):
		img_array=imread(os.path.join(path,img))
		img_resized=resize(img_array,(256,256,3))
		flat_data_arr.append(img_resized.flatten())
		target_arr.append(Categories.index(i))
	print(f'loaded category:{i} successfully')
flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
```

```python
#dataframe
df=pd.DataFrame(flat_data)
df['Target']=target
df.shape

```

```python

#input data
x=df.iloc[:,:-1]
#output data
y=df.iloc[:,-1]
```

```python
# Splitting the data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,
											random_state=10,
											stratify=y)
```

```python
x_train.shape, x_test.shape
```

```python
# Creating a support vector classifier
sv=svm.SVC()
```

```python
# Training the model using the training data
sv.fit(x_train,y_train)
```

```python
# Testing the model using the testing data
y_pred = sv.predict(x_test)

# Calculating the accuracy of the model
accuracy = accuracy_score(y_pred, y_test)

# Print the accuracy of the model
print(f"The model is {accuracy*100}% accurate")
```

```python
print(classification_report(y_test, y_pred, target_names=['no', 'yes']))


# Calculating the specificity,sensitivity of the model
specificity=recall_score(y_test, y_pred)
print(specificity)
```
